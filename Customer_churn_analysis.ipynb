Correlations between customer data features and customer churn

To decide which features of the data to include in our predictive churn model, 
weâ€™ll examine the correlation between churn and each customer feature.

```python
corr=data.corr()
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values,annot=True,annot_kws={'size':12})
heat_map=plt.gcf()
heat_map.set_size_inches(20,15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.33,random_state=47)
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
train_y=train_y.reshape(-1,1)
train_y.shape
```

```python
test_y=test_y.reshape(-1,1)
test_y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report

lg=LogisticRegression()

lg.fit(train_x,train_y)

lg.score(train_x,train_y)
```

```python
pred=lg.predict(test_x)
pred
```

```python
print("Confusion Matrix")

confusion_matrix(test_y,pred)
```

```python
print("Classification Report")
classification_report(test_y,pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

randomForest = RandomForestClassifier(n_estimators=300 , random_state=50)
randomForest.fit(train_x, train_y)
print("Accuracy Score =",(randomForest.score(test_x, test_y)))
```
