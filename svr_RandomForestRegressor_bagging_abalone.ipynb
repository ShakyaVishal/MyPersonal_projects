```python
import pandas as pd
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data',names=['Gender','Length','Diameter','Height','Whole weight','Shucked weight','Viscera Weight','Shell weight','Age_rings'])
```

```python

#https://medium.com/coinmonks/support-vector-regression-or-svr-8eb3acf6d0ff
df
```

```python
df.head()
```

```python
df=df.drop(['Gender'],axis=1)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Shucked weight'],df['Age_rings'])
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,1:-1]
```

```python
x.head()
```

```python
y=df.iloc[:,-1]
```

```python
y.head()
```

```python
print(x.shape)
print(y.shape)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.33,random_state=50)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.svm import SVR
regressor_l=SVR(kernel='linear',degree=1)
regressor_l.fit(xtrain,ytrain)
```

```python
pred=regressor_l.predict(xtest)
```

```python
print(regressor_l.score(xtest,ytest))
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(ytest,pred))
```

```python
regressor_rbf=SVR(kernel='rbf',epsilon=1.0)
```

```python
regressor_rbf.fit(xtrain,ytrain)
```

```python
pred=regressor_rbf.predict(xtest)
print(regressor_rbf.score(xtest,ytest))
print(r2_score(ytest,pred))
```

```python
regressor_poly=SVR(kernel='poly',epsilon=1.0)

regressor_poly.fit(xtrain,ytrain)
pred=regressor_poly.predict(xtest)
print(regressor_poly.score(xtest,ytest))
print(r2_score(ytest,pred))
```

```python
import numpy as np

g=np.array([0.500,0.380,0.125,0.5770,0.2690,0.1265]).reshape(1,-1)
print(g.shape)

pred=regressor_l.predict(g)
pred
```

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor_l, x, y, cv=5)
print(scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor_rbf, x, y, cv=5)
print(scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(regressor_poly, x, y, cv=5)
print(scores)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(xtrain,ytrain)
print(lr.score(xtrain,ytrain))
pred=lr.predict(xtest)
print(pred)
```
